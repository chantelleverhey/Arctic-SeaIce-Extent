# NSIDC Sea-Ice Extent Time Series 1978-2023

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('Sea_Ice_Index_Monthly_Data_by_Year_G02135_v3.0-extent.csv')
```

```python
df.head()
```

```python
# Codeblock 4: Plot the time series graph - REMINDER: Jupyter notebook is case sensitive
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Annual'], marker='o', linestyle='-')
plt.title('Sea Ice Outlook 1978-2023 (Pan-arctic region)')
plt.xlabel('Year')
plt.ylabel('Sea Ice Extent (million square km)')
plt.grid(True)

# Fit a line of best fit to the data
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year'], df['Annual'])
line = slope * df['Year'] + intercept

# Plot the line of best fit
plt.plot(df['Year'], line, color='red', label='Line of Best Fit')

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Year' column to datetime format
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

# Plot the sea-ice extent for each month
months = ['August','September','October', 'November', 'December', 'January', 'February', 'March', 'April', 'May', 'June',]

plt.figure(figsize=(12, 8))
for month in months:
    plt.plot(df['Year'], df[month], label=month, marker='o')

plt.title('Monthly Sea-Ice Extent 1978-2023')
plt.xlabel('Year')
plt.ylabel('Extent (million square km)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
